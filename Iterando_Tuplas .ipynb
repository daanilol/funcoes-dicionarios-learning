# Built-in-Function 
*zip()*

```python
#zipando lista

carros = ['Jetta Variant', 'Passat', 'Crossfox', 'DS5']
carros
```

```python
valores = [88078.64, 106161.94, 72832.16, 123549.07]
valores
```

```python
#iterando nome e valor do veículo

list(zip(carros, valores))
```

```python
#zipando, empacotando
zip(carros, valores)
```

```python
#teste
for item in zip(carros, valores):
  print(item)
```

```python
#elaborando filtragem
for carros, valores in zip(carros, valores):
  if valores > 100000:
    print(carros)
```
