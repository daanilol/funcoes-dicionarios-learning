# Iterando em dicionários

```python
dados = {
    'Jetta Variante': 88078.64, 
    'Passat': 106161.94, 
    'Crossfox': 72832.16, 
    'DS5': 124549.07, 
    'Fusca': 60589.87
    }
```

# *dict(keys)*

Retorna uma lista com todas as chaves do dicionário.

```python
dados.keys()
```

```python
for key in dados.keys():
  print(dados[key])
```

# *dict.values()*

Retorna uma lista com todas as chaves do dicionário.

```python
dados.values()
```

# *dict.item()*

Retorna uma lista contendo uma tupla para cada (chave-valor) do dicionário.

```python
dados.items()
```

```python
for item in dados.items():
  print(item)
```

```python
for key, value in dados.items():
  print(key, value)
```

```python
for key, value in dados.items():
  if value > 100000:
    print(key, value)
```
