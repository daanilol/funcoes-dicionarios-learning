```python
carros = ['Jetta Variante', 'Passat', 'Crossfox']
carros
```

```python
valores = [88078.64, 106161.94, 72832.16]
valores
```

```python
carros.index('Crossfox')
```

```python
valores[carros.index('Crossfox')]
```

```python
#criando um dicionario
dados = {'Jetta Variante': 88078.64, 'Passat': 106161.94, 'Crossfox': 72832.16}
```

```python
type(dados)
```

# Criando dicionário com *zip()*

```python
#criando dicionários com zip
dados = dict(zip(carros, valores))
```

# *dict[key]*

```python
dados['Passat']
```

# *len(dict)*

```python
len(dados)
```

# *dict[key] = value*

```python
#criando uma nova variavel dentro do dicionario
dados['DS5'] = 12459.07
```
