# *dict.update()*
Atualiza o dicionário.

```python
#criando dicionario
dados = {'Jetta Variante': 88078.64, 'Passat': 106161.94, 'Crossfox': 72832.16}
```

```python
# atualizando dicionario
dados.update({'DS5': 124549.07})
dados
```

```python
dados.update({'Fusca': 60589.87})
dados
```

```python
len(dados)
```

# *dict.copy()*

```python
# gerando uma copia dos dados (sem afetar a origem)
dados_copy = dados.copy()
```

```python
dados_copy
```

```python
#deletando e checando os dicinarios
del dados_copy['Fusca']
```

```python
dados
```

```python
dados_copy
```

# *dict.pop(key[,default])*

Deleta e retorna o valor da chave deletada.

```python
dados_copy.pop('Passat')
```

```python
dados_copy
```

```python
#caso queira deletar uma chave que não existe, ele enviará um erro
#tratando o erro a ser enviado e [default]^

'''dados_copy.pop('Passat', 'Chave não encontrada')'''
```

```python
dados_copy.pop('DS5')
```

```python
dados_copy
```

# *dict.clear()*

Apagando o os dados do dicionário.

```python
dados_copy.clear()
dados_copy
```
