# Funções com e sem parametros

Funções são unidades de códigos que realizam uma tarefa específica, podem receber alguma entrada e também podem retornar algum resultado.

```python
#sem parametros
#media 

def media():
  valor = (5 + 5 + 5 + 5) / 4
  print(valor)
media()
```

#

```python
#com parametros
#media

def media(number_1, number_2, numer_3, number_4):
  valor = (number_1 + number_2 + numer_3 + number_4) / 4
  print(valor)
media(10, 5, 10, 5)
```

```python
#com lista
#media

def media(lista):
  valor = sum(lista) / len(lista)
  print(valor)
media([10, 5, 10, 5])
```
